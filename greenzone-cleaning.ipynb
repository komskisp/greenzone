```python
import pandas as pd
```

```python
read_file = pd.read_excel (r'/Users/pauliuskomskis/Downloads/wisconsin.xls')
read_file.to_csv (r'/Users/pauliuskomskis/Desktop/github/greenzone/data/2019/west_virginia.csv')
df = pd.read_csv('/Users/pauliuskomskis/Desktop/github/greenzone/data/2019/west_virginia.csv')
print('Excel to CSV conversion Success')
df
df.columns = df.iloc[3] # taking 3rd row and making it column headers
df.drop(df.index[3]) # deleting the 3rd row from df
df = df.dropna() # dropping NaN value rows/data
df.drop(df.iloc[:,:1], inplace = True, axis = 1) # droping the firs row as it is a duplicate of headers
df.set_index('City')
df = df.drop(df.index[0]) # dropping a duplicate column
df.reset_index(drop=True)
```

```python
df.rename(columns={'City': 'city',
                  'Population': 'population',
                  'Violent\ncrime': 'violent_crime',
                  'Murder and\nnonnegligent\nmanslaughter': 'murder',
                  'Rape1': 'rape',
                  'Robbery': 'robbery',
                  'Aggravated\nassault': 'assault',
                  'Property\ncrime': 'property_crime',
                  'Burglary': 'burglary',
                  'Larceny-\ntheft': 'larceny',
                  'Motor\nvehicle\ntheft': 'gta',
                  'Arson': 'arson'}, inplace=True)
df.reset_index(drop=True, inplace=True)
```
